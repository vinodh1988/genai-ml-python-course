```python
#in python there is no support for function overloading directly

def perform():
    print("Performing default operation.")

def perform(action):
    print(f"Performing action: {action}") # existing function is overridden

#perform()  # This will raise an error
perform("Jump")  # This will work
```

### function with default parameter

#### this can be used to emulate overloading indirectly

```python
def perform(action="Run"):
    print(f"Performing action: {action}")

perform()  # This will use the default parameter
perform("Jump")  # This will override the default parameter
```

```python
def process_person(name, age=30, city="Unknown"):
    print(f"Name: {name}, Age: {age}, City: {city}")

process_person("John", 25)
process_person("Alice", 30, "New York")
process_person(name="Bob", age=22)
process_person(age="28", name="Eve", city="Los Angeles")
process_person("Charlie", "Chennai")  # uses all default values except name
#process_person(city="Lose Angeles") not allowed
```

```python
def add(a: int, b: int) -> int:
    return a + b

def sum(a: float, b: float) -> float:
    return a + b

print(add(5, 10))
print(sum(5.5, 10.2))
print(add(12.24,3.35))
print(sum(12, 3))
```

```python
# function with variable number of arguments
def add(*args): #args is a tuple, your input parameters will be packed into a tuple
    total = 0
    for num in args:
        total += num
    return total

print(add(1, 2, 3))
print(add(10, 20, 30, 40, 50))
print(add(5))
print(add())  # No arguments
print(add(1,2,3))
```

### Functions as First-Class Citizens in Python

- Functions can be assigned to variables.
- Functions can be passed as arguments to other functions.
- Functions can be returned from other functions.
- Functions can be stored in data structures (lists, dictionaries, etc.).

This allows for flexible and powerful programming patterns such as higher-order functions, callbacks, and decorators.

```python
# Function can be asssigned to variables
def greet():
    return "Hello, World!"

sayHello = greet  # Assigning function to a variable
print(sayHello())  # Calling the function using the new variable
```

```python
# functions can be stored in data structures like lists or dictionaries
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    return a / b

listops = [add, subtract, multiply, divide]

#listops[0](10, 5)  # Calls add(10, 5)
print(listops[0](10, 5))  # Calls add(10, 5)
print(listops[1](10, 5))  # Calls subtract(10, 5)
print(listops[2](10, 5))  # Calls multiply(10, 5)
print(listops[3](10, 5))  # Calls divide(10, 5)
```
