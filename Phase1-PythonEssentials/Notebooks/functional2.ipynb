```python
listx=[43,356,23,89,12,5,67,34,90]

def maniplist(listx,func): # first parameter is list, second is function
    result=[]
    for i in listx:
        result.append(func(i)) # func is a function which carries one parameter and returns something
    return result
```

```python
def plus5(x):
    return x+5
```

```python
def nearest13(x):
    lower = x
    upper = x
    while lower % 13 != 0 and upper % 13 != 0:
        lower -= 1
        upper += 1
    if lower % 13 == 0:
        return lower
    else:
        return upper
```

```python
result1=maniplist(listx,plus5)
result2=maniplist(listx,nearest13)
result3=maniplist(listx,lambda x: "Five Multiple" if x%5==0 else x)
```

```python
print(f"result1 for plus5: {result1}")
print(f"result2 for nearest13: {result2}")
print(f"result3 for lambda five multiple: {result3}")
```

```python
def myfilter(listx,func):
    result=[]
    for i in listx:
        if func(i):
            result.append(i)
    return result
```

```python
def evencheck(x):
    return x%2==0
```

```python
filteresult=myfilter(listx,evencheck)
print(f"Filtered even numbers: {filteresult}")
filteresult_lambda=myfilter(listx,lambda x: x>50)
print(f"Filtered numbers greater than 50: {filteresult_lambda}")
```
