### Python Literals

* **Literals** are data given in a variable or constant.
* Python supports several types of literals:
    - **Numeric literals**: Integers (`a = 20`), floating-point numbers (`b = 40.4`)
    - **String literals**: Text enclosed in single or double quotes (`d = "Hello World"`)
    - **Boolean literals**: `True` and `False` (`c = True`)
* Literals represent fixed values and are assigned directly to variables.
* Python automatically infers the type of literal assigned to a variable.

* Python also supports **binary**, **octal**, and **hexadecimal** literals:
    - **Binary**: Prefix with `0b` or `0B` (e.g., `x = 0b1010`)
    - **Octal**: Prefix with `0o` or `0O` (e.g., `y = 0o12`)
    - **Hexadecimal**: Prefix with `0x` or `0X` (e.g., `z = 0xA`)
* These literals are useful for working with different number systems in programming.

````python
print(a)
a=0b1010  # binary representation of 10
print(a)
a=0o12    # octal representation of 10
print(a)
a=0xA     # hexadecimal representation of 10
print(a)
````

## Formatted output in python

Formatted output in Python allows you to display variables and values in a structured and readable way. There are several methods for formatting output:

- **Using the `print()` function with commas:**  
    This automatically adds spaces between items.  
    Example:  
    `print("The value of a is", a)`

- **String concatenation:**  
    Combine strings and variables using `+`.  
    Example:  
    `print("The value of b is " + str(b))`

- **Old-style formatting with `%`:**  
    Use format specifiers like `%d` for integers, `%f` for floats, `%s` for strings.  
    Example:  
    `print("a = %d, b = %.2f, d = %s" % (a, b, d))`

- **`str.format()` method:**  
    Insert values into placeholders `{}`.  
    Example:  
    `print("a = {}, b = {:.2f}, d = {}".format(a, b, d))`

- **f-strings (Python 3.6+):**  
    Embed expressions inside string literals, prefixed with `f`.  
    Example:  
    `print(f"a = {a}, b = {b:.2f}, d = {d}")`

Formatting options allow control over decimal places, alignment, padding, and more. F-strings are the most modern and recommended approach for formatted output in Python.

# Python collections

Python collections are data structures used to store multiple values. The main built-in collection types are:

- **List**: Ordered, mutable sequence of elements. Example: `my_list = [1, 2, 3]`
- **Tuple**: Ordered, immutable sequence of elements. Example: `my_tuple = (1, 2, 3)`
- **Set**: Unordered collection of unique elements. Example: `my_set = {1, 2, 3}`
- **Dictionary**: Unordered collection of key-value pairs. Example: `my_dict = {'a': 1, 'b': 2}`

Collections allow you to group related data and perform operations such as iteration, searching, and modification. Lists and dictionaries are the most commonly used collection types in Python.

# Set

* Set is an unordered collection of unique elements
* Sets are not indexed (no position-based access)
* Elements are not stored in any particular order
* You cannot access or change elements by index
* Sets do not allow duplicate values
* Sets can contain elements of different data types
* Set elements are enclosed within `{}` (curly braces)
* Useful for membership testing, removing duplicates, and set operations (union, intersection, difference)

````python
seta={1, 2, 3, 4, 5, 5, 4, 3}
print(seta)  # duplicates removed
setb={"apple", "banana", "cherry", "apple", 1,2,3,2,1}
print(setb)  # duplicates removed
seta.add(6)
seta.remove(3)
````

## Set operations

Set operations allow you to perform mathematical operations on sets, such as combining, comparing, and finding differences between them. The main set operations in Python are:

- **Union (`seta.union(setb)` or `seta | setb`)**: Combines all unique elements from both sets.
- **Intersection (`seta.intersection(setb)` or `seta & setb`)**: Returns elements common to both sets.
- **Difference (`seta.difference(setb)` or `seta - setb`)**: Returns elements in the first set but not in the second.
- **Symmetric Difference (`seta.symmetric_difference(setb)` or `seta ^ setb`)**: Returns elements in either set, but not in both.

These operations are useful for tasks like removing duplicates, finding common elements, and comparing datasets.

````python
for item in seta:
    print(item)

seta={1, 2, 3, 4, 5}
setb={4, 5, 6, 7, 8,1,3}
print("Union:", seta.union(setb))
print("Intersection:", seta.intersection(setb))
print("Difference (seta - setb):", seta.difference(setb))
print("Difference (setb - seta):", setb.difference(seta))
print("Symmetric Difference:", seta.symmetric_difference(setb))
````

### Dictionary in python

A **dictionary** in Python is an unordered collection of key-value pairs. Each key must be unique and immutable (such as strings, numbers, or tuples), while values can be of any data type and can be duplicated. Dictionaries are defined using curly braces `{}` with keys and values separated by a colon `:`.

**Example:**
```python
my_dict = {'name': 'Alice', 'age': 25, 'city': 'New York'}
```

**Key features:**
- Access values using keys: `my_dict['name']` returns `'Alice'`
- Add or update values: `my_dict['age'] = 26`
- Remove items: `del my_dict['city']`
- Check if a key exists: `'name' in my_dict`
- Iterate over keys, values, or items:
    - `for key in my_dict:`
    - `for value in my_dict.values():`
    - `for key, value in my_dict.items():`

Dictionaries are useful for representing structured data and for fast lookups by key.

````python
person = {
    "name": "John",
    "age": 30,
    "city": "New York"
}

print("Name:", person["name"])
print("Age:", person["age"])
print("City:", person["city"])

for key, value in person.items():
    print(f"{key}: {value}")

for key in person.keys():
    print(key)

for value in person.values():
    print(value)

````
