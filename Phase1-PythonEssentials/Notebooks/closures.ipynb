#Closures in python

A closure is a function object that remembers values in enclosing scopes even if those scopes are no longer present. Closures are used to retain state and encapsulate behavior.

**Example:**
```python
def outer_func(msg):
    def inner_func():
        print(msg)
    return inner_func

closure = outer_func("Hello, Closure!")
closure()  # Output: Hello, Closure!
```

**Key points:**
- The inner function (`inner_func`) remembers the value of `msg` from the enclosing scope.
- Closures are useful for callback functions, decorators, and maintaining state.
