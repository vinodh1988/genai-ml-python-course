```python
listx=(14,534,34,53,23,12,65,78,91,26,39,48,50,61,73,88,99,100)

result =map(lambda x: x**2,listx)
def plus5(x):
    return x+5
result2 = map(plus5,listx)
print(f"result1: {list(result)}")
print(f"result2: {list(result2)}")
```

```python
listnames=["Alice","Bob","Charlie","David","Eve","Frank","Grace","Heidi","Harrison","Johnson"]
result_names=map(lambda x: x.upper(),listnames)
print(f"Uppercase names: {list(result_names)}")
```

```python
filterednames=filter(lambda x: len(x)>4,listnames)
print(f"Filtered names with length >4: {list(filterednames)}")

def filterString(name):
    return name.startswith('H') or name.startswith('M') or name.startswith('J')

filterednames2=filter(filterString,listnames)
print(f"Filtered names starting with H, M, J: {list(filterednames2)}")
```

```python
import functools

listx=(14,534,34,53,23,12,65,78,91,26,39,48,50,61,73,88,99,100)

result = functools.reduce(lambda x,y: x+y,listx)

print(f"Sum of all elements in listx: {result}")

def sumofEven(x,y):
    if y%2==0:
        return x+y
    else:
        return x

result2 = functools.reduce(sumofEven,listx)
print(f"Sum of even elements in listx: {result2}")
```
